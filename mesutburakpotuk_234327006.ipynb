# 4- Görüntüye unsharp masking ve highboost filtering uygulama, görüntüyü ve histogramını çizdirme

"unsharp masking" işlemi için kullanılan ağırlık katsayısı (alpha), genellikle 1 ile 2 arasında bir değer alır. Bu değer, orijinal görüntüye ne kadar vurgu yapılacağını belirler.

```python
# ALPHA DEĞERİNİ 1 OLARAK DENEDİM. ÇIKAN HİSTOGRAM GRAFİĞİ HEDEFLENENE BENZEMEDİ.


import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "resim.jpeg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

equ = cv2.equalizeHist(img)

resized_img = cv2.resize(equ, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

blurred = cv2.GaussianBlur(resized_img, (0, 0), 3)
unsharp_img = cv2.addWeighted(resized_img, 1, blurred, -0.1, 0)

k_highboost = 0.1
highboost_img = cv2.addWeighted(resized_img, k_highboost, unsharp_img, -k_highboost, 0)

plt.subplot(4, 2, 7)
plt.imshow(unsharp_img, cmap='gray')
plt.title('Highboost Image')
plt.axis('off')

plt.subplot(4, 2, 8)
plt.hist(unsharp_img.ravel(), 256, [0, 256])
plt.title('Highboost Histogram')

plt.tight_layout()
plt.show()
```

# 5- Highboost sonucunun dft2'sini alma dft2'yi shift etme, görüntüyü ve genlik spektrumunu çizdirme

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "resim.jpeg"
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

equ = cv2.equalizeHist(img)

resized_img = cv2.resize(equ, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

blurred = cv2.GaussianBlur(resized_img, (0, 0), 3)
unsharp_img = cv2.addWeighted(resized_img, 1.8, blurred, -0.1, 0)

k_highboost = 0.1
highboost_img = cv2.addWeighted(resized_img, k_highboost, unsharp_img, -k_highboost, 0)

dft2_highboost = np.fft.fft2(unsharp_img)
dft2_shifted_highboost = np.fft.fftshift(dft2_highboost)
magnitude_spectrum_highboost = np.log(np.abs(dft2_shifted_highboost) + 1)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(unsharp_img, cmap='gray')
plt.title('Highboost Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(magnitude_spectrum_highboost, cmap='gray')
plt.title('Highboost Magnitude Spectrum')
plt.axis('off')

plt.tight_layout()
plt.show()
```
